```python
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
base_path = Path("runs/benchmark-augmentation")

for run_dir in base_path.glob("*/run0"):
    plt.figure(figsize=(16, 5))
    plt.suptitle(run_dir.parts[-2])
    gs = plt.GridSpec(1, 3)
    for shuffle in sorted(run_dir.glob('shuffle*/')):
        if not (shuffle / 'train.csv').exists():
            continue
        
        ax = plt.subplot(gs[0])
        df = pd.read_csv(shuffle / 'train.csv')
        plt.plot(df.epoch, df.loss, label=shuffle.parts[-1])
        plt.legend()
        
        ax = plt.subplot(gs[1])
        df = pd.read_csv(shuffle / 'train.csv')
        plt.plot(df.epoch, df.accuracy, label=shuffle.parts[-1])
        plt.ylim(0, 1)
        plt.legend()
        
        ax = plt.subplot(gs[2])
        df = pd.read_csv(shuffle / 'val.csv')
        plt.plot(df.epoch, df.accuracy, label=shuffle.parts[-1])
        plt.ylim(0, 1)
        plt.legend()
        
    plt.show()
```

## Compare Means

```python
base_path = Path("runs/benchmark-augmentation")

plt.figure(figsize=(16, 5))
gs = plt.GridSpec(1, 3)

for run_dir in base_path.glob("*/run0"):
    train_dfs = []
    val_dfs = []
    for shuffle in sorted(run_dir.glob('shuffle*/')):
        if not (shuffle / 'train.csv').exists():
            continue
        
        df = pd.read_csv(shuffle / 'train.csv')
        # df['shuffle'] = shuffle.parts[-1]
        train_dfs.append(df)
        
        df = pd.read_csv(shuffle / 'val.csv')
        # df['shuffle'] = shuffle.parts[-1]
        val_dfs.append(df)
        
    df = pd.concat(train_dfs)
    df = df.reset_index()
    
    ax = plt.subplot(gs[0])
    plt.title("Training Loss")
    sns.lineplot(data=df, x='epoch', y='loss', ax=ax, label=run_dir.parts[-2])
    plt.legend()
    
    ax = plt.subplot(gs[1])
    plt.title("Training Accuracy")
    sns.lineplot(data=df, x='epoch', y='accuracy', ax=ax, label=run_dir.parts[-2])
    plt.legend()
    
    df = pd.concat(val_dfs)
    df = df.reset_index()
    ax = plt.subplot(gs[2])
    plt.title("Test Accuracy")
    sns.lineplot(data=df, x='epoch', y='accuracy', ax=ax, label=run_dir.parts[-2])
    plt.legend()
```
